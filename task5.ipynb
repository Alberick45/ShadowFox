```python
class MobilePhone:
    def __init__(self,screen_type="Touch Screen",Network_type=None, dualsim=None, frontcamera=None,rearcamera=None,Ram=None,Storage=None):
        valid_screen_type=["Touch Screen"]
        valid_Network_type=["4G","5G"]
        valid_dualsim=[True,False] 
        valid_frontcamera=["5MP","8MP","12MP","16MP"]
        valid_rearcamera=["8MP","12MP","16MP","32MP","48MP"]
        valid_Ram=["2GB","3GB","4GB"]
        valid_Storage=["16GB","32GB","64GB"]

        if screen_type in valid_screen_type:
            self.screen_type = screen_type
        if Network_type in valid_Network_type:    
            self.Network_type = Network_type
        if dualsim in valid_dualsim:
            self.dualsim = dualsim
        if frontcamera in valid_frontcamera:
            self.frontcamera = frontcamera
        if rearcamera in valid_rearcamera:
            self.rearcamera = rearcamera
        if Ram in valid_Ram:
            self.Ram = Ram
        if Storage in valid_Storage:
            self.Storage = Storage
        pass
    def make_call(self):
        print("Making call")
    def recieve_call(self):
        print("recieving call")
    def take_picture(self):
        print("taking a picture")
    def send_message(self):
        print("sending message")
    def recieve_message(self):
        print("recieving message")
    def browse(self):
        print("browsing the internet")
    5
class Samsung(MobilePhone):                            
    def __init__(self,screen_type,Network_type, dualsim, frontcamera,rearcamera,Ram,Storage,bixby,one_ui=True,samsung_pay=True,Os="android",s_pen=True):
        super().__init__(
            screen_type,Network_type,dualsim,frontcamera,rearcamera,Ram,Storage
        )
        self.bixby = bixby
        self.one_ui = one_ui
        self.samsung_pay = samsung_pay
        self.Os = Os
        self.s_pen = s_pen
        
class Apple(MobilePhone):
    def __init__(self,screen_type,Network_type, dualsim, frontcamera,rearcamera,Ram,Storage,siri,icloud=True,apple_pay=True,Os="ios"):
        super().__init__(
            screen_type,Network_type,dualsim,frontcamera,rearcamera,Ram,Storage
        )
        self.siri = siri
        self.icloud = icloud
        self.apple_pay = apple_pay
        self.Os = Os
        
    
```
